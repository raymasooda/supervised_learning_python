# Split the Data

```python
## Let's look at the distribution of targets ##
y = data[[target]]

## Plot histogram ##
plt.figure(figsize=(3,2))
plt.hist(y, bins=20,alpha=0.7)

## Set axes ##
counts = y.value_counts()
plt.xticks([0,1],[f'{label} ({count})' for label,count in zip(['Did not donate','Did donate'], counts)])
plt.ylabel('number of people');

prop_donated = round((counts[1] / (counts[0]+counts[1])),3)
print(f'{prop_donated*100}% of people donated in March 2007')
```

- Yikes, that's pretty imbalanced. Might be worth using some imblearn techniques in the future (e.g. random oversampling) but will stick with just stratifying the data splitting, and use class_weight parameter of sklearn classifiers to penalize mistakes on the different classes by their weight.
- Since the smaller class has enough samples, will use a relatively high kfold cross-validation of 5 folds

**7.1 Data splitting code sample**

```python
## First split data into held-out (i.e. test) data, and model selection (i.e. train + validate) data ##
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y, test_size=0.2, random_state=42)

## Define cross-validation splits for model selection ##
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
```

# Define the Model

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

# Fit the model #
First evaluate the training model (i.e. calculate accuracy on training/validation data used in model selection process) and check for overfitting, using the following three models:
1. Fit model with no cross-validation or hyperparameter optimization (i.e. **unvalidated_model**)
2. Fit model with cross-validation, but no hyperparameter optimization (i.e. **validated_model**)
3. Fit model with both cross-validation and hyperparameter optimization (i.e. **validated_tuned_model**)

```python
## No cross-validation or hyperparameter optimization ##
unvalidated_model = clf.fit(X_train, y_train)

## Cross-validated, but no hyperparameter optimization ##
validated_predictions, validated_scores, validated_model = fn.pipeline(clf, skf, X_train.to_numpy(), y_train.to_numpy())
```

```python
from sklearn.model_selection import cross_val_predict
## Set up figure parameters ##
fig, ax = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,3), sharey=True)
cols = {'precision':0, #precision subplot at column 1
       'recall':1, #recall subplot at column 2
       'accuracy':2} #accuracy subplot at column 3

## Plot accuracy scores for training/testing sets for the validated model first ##
metric_list = list(set([y for x in validated_scores.columns for y in x.split('_') if ('train' not in y) & ('validate' not in y)]))
for metric in metric_list:
    col = cols[metric]
    metric_df = validated_scores.T[validated_scores.T.index.str.contains(metric)].T
    fn.plot_metric(metric_df, metric, 'validate', ax[col], col)       
    
    ## Plot metrics of unvalidated model ##
    unvalidated_score = fn.generate_score(unvalidated_model.predict(X_train), y_train, metric)
    ax[col].axhline(unvalidated_score, color='grey', ls='--', label='No CV')
    
    ax[col].legend() if col == 2 else None
```

- Lots of overfitting happening, as the validation split (blue) is much lower than the train split (red), as well as the data set with no cross-validation (black).
- Both precision and recall are quite bad, where it seems like there is not enough positive cases for the model to learn these positive cases (i.e. lower recall on the training set). In contrast, precision is nearly perfect on the training data since there are lots of negative cases. Maybe the validation splits are too small?

**Test whether weighting the classes in the classifier model can balance out the recall and precision scores a bit more**

```python
from sklearn.utils.class_weight import compute_class_weight

## Calculate class weights based on the class distribution in your dataset ##
class_weights = compute_class_weight('balanced', classes=np.unique(list(y_train.iloc[:,0])), y=list(y_train.iloc[:,0]))

## Create a dictionary of class weights ##
class_weight_dict = dict(enumerate(class_weights))

## Define model with weighted classes ##
clf_weighted = DecisionTreeClassifier(class_weight=class_weight_dict)
```

```python
## Set up figure parameters ##
fig, ax = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,3), sharey=True)
cols = {'precision':0, #precision subplot at column 1
       'recall':1, #recall subplot at column 2
       'accuracy':2} #accuracy subplot at column 3

## Plot accuracy scores for training/testing sets for the validated model first ##
metric_list = list(set([y for x in validated_scores_weighted.columns for y in x.split('_') if ('train' not in y) & ('validate' not in y)]))
for metric in metric_list:
    col = cols[metric]
    metric_df = validated_scores_weighted.T[validated_scores_weighted.T.index.str.contains(metric)].T
    fn.plot_metric(metric_df, metric, 'validate', ax[col], col)    
plt.ylim(0,1)
fig.suptitle('Weighted classes', y=1.08);
```

- The recall definitely increased, but maybe the weighting was slightly too strong (since the precision dipped a fair amount). But definitely looks more balanced than before.

**Now try and improve both scores with some hyperparameter optimization, using randomized search CV**
- Here, we are specifically using the class-weighted classifier

```python
## Model with cross-validated and hyperparameter tuning using RandomizedSearchCV ##
params = {'max_depth':range(2,50),
          'min_samples_split':range(2,100),
          'criterion':['gini','entropy'],
          'min_impurity_decrease':np.linspace(0,1,21)}
randomCV = RandomizedSearchCV(estimator=clf_weighted, 
                              param_distributions=params, 
                              cv=5, #stratified K-fold the default for binary classification
                              scoring='accuracy',
                              n_iter=20,
                              random_state=42)
randomCV.fit(X_train, y_train)
validated_tuned_model = randomCV.best_estimator_
```

```python
## Set up figure parameters ##
fig, ax = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,3), sharey=True)
cols = {'precision':0, #precision subplot at column 1
       'recall':1, #recall subplot at column 2
       'accuracy':2} #accuracy subplot at column 3

## Plot accuracy scores for training/testing sets for the validated model first ##
metric_list = list(set([y for x in validated_scores_weighted.columns for y in x.split('_') if ('train' not in y) & ('validate' not in y)]))
for metric in metric_list:
    col = cols[metric]
    metric_df = validated_scores_weighted.T[validated_scores_weighted.T.index.str.contains(metric)].T
    fn.plot_metric(metric_df, metric, 'validate', ax[col], col)       
    
    ## Plot metrics of best validated + tuned model ##
    tuned_score = fn.generate_score(cross_val_predict(validated_tuned_model, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score, color='mediumseagreen', ls='--', label='randCV best model (validate)')
    ax[col].legend() if col == 2 else None
fig.suptitle('Weighted classes', y=1.08);
```

- Doing okay on precision, but still struggling with recall.
- There are lots of parameters so maybe 20 iterations of the randomzied search CV is not enough?

**Could a simple increasing of iteration of the randomized search improve the recall (and the precision, for that matter)?**
- Test 200 iterations, instead of 20

```python
randomCV_100 = RandomizedSearchCV(estimator=clf_weighted, 
                              param_distributions=params, 
                              cv=5, #stratified K-fold the default for binary classification
                              scoring='accuracy',
                              n_iter=100,
                              random_state=42)
randomCV_100.fit(X_train, y_train)
validated_tuned_model_100 = randomCV_100.best_estimator_
```

```python
## Set up figure parameters ##
fig, ax = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,3), sharey=True)
cols = {'precision':0, #precision subplot at column 1
       'recall':1, #recall subplot at column 2
       'accuracy':2} #accuracy subplot at column 3

## Plot accuracy scores for training/testing sets for the validated model first ##
metric_list = list(set([y for x in validated_scores_weighted.columns for y in x.split('_') if ('train' not in y) & ('validate' not in y)]))
for metric in metric_list:
    col = cols[metric]
    metric_df = validated_scores_weighted.T[validated_scores_weighted.T.index.str.contains(metric)].T
    fn.plot_metric(metric_df, metric, 'validate', ax[col], col)       
    
    ## Plot metrics of best validated + tuned model with 20 iterations ##
    tuned_score = fn.generate_score(cross_val_predict(validated_tuned_model, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score, color='mediumseagreen', ls='--', label='randCV20 best model (validate)')
    
    ## Plot metrics of best validated + tuned model with 100 iterations ##
    tuned_score_100 = fn.generate_score(cross_val_predict(validated_tuned_model_100, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score_100, color='orange', ls='--', label='randCV100 best model (validate)')
    
    ax[col].legend() if col == 2 else None
fig.suptitle('Weighted classes', y=1.08);
```

- Only a marginal increase in the recall, so maybe not the best approach.
- What about changing the scoring metric to recall, instead of accuracy for the randomized search CV?

**Use recall score as the scoring criteria for randomized search CV**
- Here, we will continue using 100 iterations

```python
randomCV_recall = RandomizedSearchCV(estimator=clf_weighted, 
                              param_distributions=params, 
                              cv=5, #stratified K-fold the default for binary classification
                              scoring='recall',
                              n_iter=100,
                              random_state=42)
randomCV_recall.fit(X_train, y_train)
validated_tuned_model_recall = randomCV_recall.best_estimator_
```

```python
## Set up figure parameters ##
fig, ax = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,3), sharey=True)
cols = {'precision':0, #precision subplot at column 1
       'recall':1, #recall subplot at column 2
       'accuracy':2} #accuracy subplot at column 3

## Plot accuracy scores for training/testing sets for the validated model first ##
metric_list = list(set([y for x in validated_scores_weighted.columns for y in x.split('_') if ('train' not in y) & ('validate' not in y)]))
for metric in metric_list:
    col = cols[metric]
    metric_df = validated_scores_weighted.T[validated_scores_weighted.T.index.str.contains(metric)].T
    fn.plot_metric(metric_df, metric, 'validate', ax[col], col)       
    
    ## Plot metrics of best validated + tuned model with 100 iterations ##
    tuned_score_100 = fn.generate_score(cross_val_predict(validated_tuned_model_100, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score_100, color='orange', ls='--', label='randCV100 best accuracy model (validate)')
    
    ## Plot metrics of best validated + tuned model with 100 iterations ##
    tuned_score_recall = fn.generate_score(cross_val_predict(validated_tuned_model_recall, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score_recall, color='magenta', ls='--', label='randCV100 best recall model (validate)')
    
    ax[col].legend() if col == 2 else None
fig.suptitle('Weighted classes', y=1.08);
```

**Implement imblearn library to oversample the minority class**
- Important to use an unweighted classifier after balancing the classes with oversampling
- Fit model to overampled data, then predict on real (i.e. not oversampled) data

```python
from imblearn.over_sampling import RandomOverSampler

# Create the RandomOverSampler object
oversampler = RandomOverSampler(sampling_strategy='auto', random_state=42)

# Fit and transform your training data
X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train, y_train)
```

```python
randomCV_100_resampled = RandomizedSearchCV(estimator=clf, #use unweighted classifier, since the oversampling re-weighted the classes
                              param_distributions=params, 
                              cv=5, #stratified K-fold the default for binary classification
                              scoring='accuracy',
                              n_iter=100,
                              random_state=42)
randomCV_100_resampled.fit(X_train_resampled, y_train_resampled)
validated_tuned_model_resampled = randomCV_100_resampled.best_estimator_
```

```python
## Set up figure parameters ##
fig, ax = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,3), sharey=True)
cols = {'precision':0, #precision subplot at column 1
       'recall':1, #recall subplot at column 2
       'accuracy':2} #accuracy subplot at column 3

## Plot accuracy scores for training/testing sets for the validated model first ##
metric_list = list(set([y for x in validated_scores_weighted.columns for y in x.split('_') if ('train' not in y) & ('validate' not in y)]))
for metric in metric_list:
    col = cols[metric]
    metric_df = validated_scores_weighted.T[validated_scores_weighted.T.index.str.contains(metric)].T
    fn.plot_metric(metric_df, metric, 'validate', ax[col], col)       
    
    ## Plot metrics of best validated + tuned model with 100 iterations ##
    tuned_score_100 = fn.generate_score(cross_val_predict(validated_tuned_model_100, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score_100, color='orange', ls='--', label='randCV100 best accuracy model (validate)')
    
    ## Plot metrics of best validated + tuned model with 100 iterations ##
    tuned_score_resampled = fn.generate_score(cross_val_predict(validated_tuned_model_resampled, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score_resampled, color='chocolate', ls='--', label='randCV100 best accuracy model resampled (validate)')
    
    ax[col].legend() if col == 2 else None
plt.ylim(0,1)
fig.suptitle('Weighted classes', y=1.08);
```

- Oversampling didn't work too well.

**Implement imblearn library to undersample the minority class**

```python
from imblearn.under_sampling import RandomUnderSampler

# Create the RandomOverSampler object
undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=42)

# Fit and transform your training data
X_train_undersampled, y_train_undersampled = undersampler.fit_resample(X_train, y_train)
```

```python
randomCV_100_undersampled = RandomizedSearchCV(estimator=clf, #use unweighted classifier, since the oversampling re-weighted the classes
                              param_distributions=params, 
                              cv=5, #stratified K-fold the default for binary classification
                              scoring='accuracy',
                              n_iter=100,
                              random_state=42)
randomCV_100_undersampled.fit(X_train_undersampled, y_train_undersampled)
validated_tuned_model_undersampled = randomCV_100_undersampled.best_estimator_
```

```python
## Set up figure parameters ##
fig, ax = plt.subplots(nrows=1, ncols=3,
                       figsize=(12,3), sharey=True)
cols = {'precision':0, #precision subplot at column 1
       'recall':1, #recall subplot at column 2
       'accuracy':2} #accuracy subplot at column 3

## Plot accuracy scores for training/testing sets for the validated model first ##
metric_list = list(set([y for x in validated_scores_weighted.columns for y in x.split('_') if ('train' not in y) & ('validate' not in y)]))
for metric in metric_list:
    col = cols[metric]
    metric_df = validated_scores_weighted.T[validated_scores_weighted.T.index.str.contains(metric)].T
    fn.plot_metric(metric_df, metric, 'validate', ax[col], col)       
    
    ## Plot metrics of best validated + tuned model with 100 iterations ##
    tuned_score_resampled = fn.generate_score(cross_val_predict(validated_tuned_model_resampled, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score_resampled, color='chocolate', ls='--', label='randCV100 best accuracy model oversampled (validate)')
    
    ## Plot metrics of best validated + tuned model with 100 iterations ##
    tuned_score_undersampled = fn.generate_score(cross_val_predict(validated_tuned_model_undersampled, X_train, y_train, cv=5), y_train, metric)
    ax[col].axhline(tuned_score_undersampled, color='darkcyan', ls='--', label='randCV100 best accuracy model undersampled (validate)')
    
    ax[col].legend() if col == 2 else None
plt.ylim(0,1)
fig.suptitle('Weighted classes', y=1.08);
```

# So is this actually solving any problem? #

- To be honest, I'm not even sure what the problem is to begin with. **Can we predict whether someone donated blood in March 2007 based on the history of their donations?** I struggle to see the point of trying to answer this question without context. If March 2007 was the most recent donation time/database update, then I guess it could be useful in predicting if someone will donate blood on the next donation, given their entire history of donations.
- The dimensionality of the data is quite low (in the feature space) so maybe there are other important factors that could be considered. There may well be no strong relationship between the current features and the target variable, and we are just trying to force a relationship onto this data, leading to overfitting/poor generalization.
- With all that being said, the amount of testing and tuning done in this notebook is not particularly extensive, and testing more hyperparameters, feature engineering, or different models might produce a different result.
