# Data reading and cleaning #
***- This section is all the same as the regression notebook (except "pre_score" is replaced by "Gender" as the target)***

**1. Upload data and check for missing values**

```python
df = pd.read_csv('MBA_ADMISSIONS.csv')
display(df.info()) #does the non-null count of every column match the index range of the dataframe?
```

**2. Split data into categorical feature matrix, numerical feature matrix, and target vector**

```python
X_df = df.drop('Gender', axis=1) #select all columns except 'Gender'
X_cat_df = X_df.T[X_df.dtypes == object].T #select all categorical columns of X
X_num_df = X_df.drop(X_cat_df.columns, axis=1) #select all numerical columns of X

y = df['Gender'].to_numpy()
y_binary = np.where(y == 'Female',1,0) #set Female as 1
```

**3. Perform one-hot encoding on categorical features and combine with numerical features**

```python
X_cat = X_cat_df.to_numpy() #transform categorical feature matrix to numpy array for sklearn

encode = OneHotEncoder() #define encoding model
encode.fit(X_cat) #fit encoding model to categorical feature array

encoded_data = encode.transform(X_cat).toarray() #transform categorical feature matrix to encoded data
encoded_column_names = encode.get_feature_names_out(X_cat_df.columns) #get new column names of encoded data
X_cat_encoded_df = pd.DataFrame(encoded_data, columns=encoded_column_names) #create a new dataframe of the encoded data (with column names)

X_pre_processed_df = pd.concat([X_num_df, X_cat_encoded_df], axis=1) #combine numerical data with encoded categorical data
X_pre_processed = X_pre_processed_df.to_numpy() #convert combined dataframe into numpy array
```

**2. Plot distributions of numerical features for males vs. females (i.e. do the distributions look different?)**

```python
## Drop all categorical columns from original data ##
categorical_columns = X_cat_df.columns
numeric_df = df.drop(categorical_columns, axis=1)
numeric_data = numeric_df.set_index('Gender') #set gender as index

## Initialize subplots ##
fig, ax = plt.subplots(nrows=1, ncols=len(numeric_data.columns),
                       figsize=(18,5),
                       sharey=True)

## Loop through each numerical feature column ##
for col, (feature, data) in enumerate(numeric_data.items()):
    xmin = data.min()
    xmax = data.max()
    bins = np.linspace(xmin, xmax, 11) #define bins so that the same bins can be used for both genders
    
    ## Loop through each gender ##
    for gender, color in zip(['Male','Female'], ['dodgerblue','indianred']):
        gender_feature_data = data[data.index == gender] #get all rows where gender = the current gender in the inner loop
        
        ## Get mean of distribution ##
        mean = round(gender_feature_data.mean(),1)
        
        ## Plot histogram with density estimate ##
        sns.histplot(gender_feature_data.to_numpy(), 
                     ax=ax[col], 
                     bins=bins, stat='probability', kde=True,
                     color=color, alpha=0.5,
                     label=f'{gender}\n({mean} avg.)')
        
    ## Set axes, legends, and titles for each subplot ##
    ax[col].set_xlabel(feature)
    ax[col].legend()
plt.show()
```

- No single numerical feature appears to discriminate between males and females that strongly, although some females tend to be higher in some features (i.e. all of the features that start with "percentage", the middle 4 subplots)
- Extremely high "post_score" tend to occur more in Males
- "Age_in_year" appears to be ordinal, which isn't surprising

**5. Seems like we were getting non-convergence in the initial model (see error message from previous cell), so try standardizing input features**

```python
## Inputs min-max normalized (i.e. between 0 and 1) ##
X_norm_df = (X_pre_processed_df - X_pre_processed_df.min()) / (X_pre_processed_df.max() - X_pre_processed_df.min())

## Inputs z-scored ##
X_zscored_df = X_pre_processed_df.apply(zscore)
```

```python
## Set up figure parameters ##
fig, ax = plt.subplots(nrows=1, ncols=6,
                       figsize=(12,3), sharey=True)
cols = {'precision':0, #precision subplots start at first column
       'recall':3} #recall subplots start at fourth column

## Set up containers for adding different models' scores and fitted models to ('raw' model already included) ##
all_scores = {'raw':raw_scores}
fitted_models = {'raw':raw_model_fit}

## Plot scores from raw input features (calculated in previous section) ##
metrics = list(set([y for x in raw_scores.columns for y in x.split('_') if ('train' not in y) & ('test' not in y)]))
for metric in metrics:
    col = cols[metric]
    metric_df = raw_scores.T[raw_scores.T.index.str.contains(metric)].T
    fn.plot_metric(metric_df, metric, 'raw', ax[col], col)
    cols[metric] += 1    

## Calculate and plot scores from normalized & z-scored input features ##
for norm, title in zip([X_norm_df, X_zscored_df],
                       ['min-max normalized','z-scored']):
    X = norm.to_numpy()
    scores, model_fit = fn.pipeline('log_reg', k_fold, X, y_binary)
    all_scores[title] = scores
    fitted_models[title] = model_fit
    
    metrics = list(set([y for x in scores.columns for y in x.split('_') if ('train' not in y) & ('test' not in y)]))
    for metric in metrics:
        col = cols[metric]
        metric_df = scores.T[scores.T.index.str.contains(metric)].T
        fn.plot_metric(metric_df, metric, title, ax[col], col)        
        
        cols[metric] += 1
```

- We are getting convergence with both min-max normalization and z-scoring of input features (i.e. no error messages)
- We are also getting beter recall (not too much difference in terms of precision) with both types of feature scaling
- I think min-max normalization makes most sense, as this is the only way to preserve the "true structure" of the one-hot-encoded categorical features (i.e. z-scoring a binary feature makes no sense, but min-max normalization won't change the values of binary features)
